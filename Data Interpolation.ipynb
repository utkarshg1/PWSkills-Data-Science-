## 1. Linear Interpolation

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([2.1,3.9,6.3,8.2,9.9])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.show()
```

```python
x_new = np.linspace(1,5,10)
y_interp = np.interp(x_new, x, y)
```

```python
y_interp
```

```python
plt.scatter(x_new, y_interp)
plt.show()
```

## 2. Cubic Interpolation

```python
x = np.array([1,2,3,4,5])
y = np.array([1.1 , 7.8, 27.5 , 63.2, 124])
```

```python
plt.scatter(x,y)
plt.show()
```

```python
from scipy.interpolate import interp1d
```

```python
# Create cubic interpolation function
f = interp1d(x,y,kind='cubic')
```

```python
x_new = np.linspace(1,5,10)
```

```python
y_interp = f(x_new)
```

```python
y_interp
```

```python
plt.scatter(x_new, y_interp)
```

## 3. Polynomial Interpolation

```python
np.random.seed(123)
x = np.array([1,2,3,4,5])
y = x**3 + np.random.normal(0,5,size=5)
```

```python
plt.scatter(x,y)
```

```python
p = np.polyfit(x,y,2)
```

```python
x_new = np.linspace(1,5,10)
y_interp = np.polyval(p, x_new)
```

```python
plt.scatter(x,y,label='actual')
plt.scatter(x_new, y_interp,label='interp')
plt.legend()
plt.show()
```

```python
p = np.polyfit(x,y,3)
x_new = np.linspace(1,5,10)
y_interp = np.polyval(p, x_new)
plt.scatter(x,y,label='actual')
plt.scatter(x_new, y_interp,label='interp3')
plt.show()
```
