# Exploratory Data Analysis (Red Wine Dataset) - Utkarsh Gaikwad 

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv'
```

```python
import pandas as pd
df = pd.read_csv(url,sep = ';')
df.head()
```

```python
# Summary of Data
df.info()
```

```python
df.shape
```

```python
# Data Descriptive Statistics
df.describe().T
```

```python
quality_counts = df['quality'].value_counts()
quality_counts
```

```python
quality_counts.plot(kind='bar',xlabel='quality',ylabel='value counts',title='Value counts for quality')
```

```python
df.columns
```

```python
df['quality'].unique()
```

```python
# Missing Values in Dataset
df.isnull().sum()
```

```python
# Checking Duplicate records
df.duplicated().sum()
```

```python
df[df.duplicated()]
```

```python
# Remove the duplicates
df = df.drop_duplicates()
```

```python
df.shape
```

```python
# Reset index
df.index = range(len(df))
```

```python
# Correlation
df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True,fmt='.3f')
```

```python
for i in df.columns:
    sns.histplot(df[i],kde=True)
    plt.title(f'Histogram for {i}')
    print('\n')
    plt.show()
```
