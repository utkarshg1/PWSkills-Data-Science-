```python
# Creating above function 
def get_duration_float(s):
    if ((s.find('h')>0) and (s.find('m')>0)):
        hrs = s.split('h')[0]
        mins = s.split(' ')[1][:-1]
    elif((s.find('h')>0) and (s.find('m')<0)):
        hrs = s.split('h')[0]
        mins = 0
    elif((s.find('h')<0) and (s.find('m')>0)):
        hrs = 0
        mins = s.split('m')[0]
    duration = float(hrs) + float(mins)/60
    return duration
```

```python
df['dur_float'] = df['Duration'].apply(get_duration_float)
```

## One Hot Encoding of Categorical Features

```python
cat_cols = list(df.select_dtypes(include='object').columns)
cat_cols
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
cat_array = ohe.fit_transform(df[cat_cols]).toarray()
cat_array
```

```python
df_cat = pd.DataFrame(cat_array,columns=ohe.get_feature_names_out())
df_cat.head()
```

```python
# Numeric features
df_num = df.select_dtypes(exclude='object')
df_num.head()
```

```python
# Join Numeric and Categorical Features
df_concat = pd.concat([df_cat,df_num],axis=1)
df_concat.head()
```
