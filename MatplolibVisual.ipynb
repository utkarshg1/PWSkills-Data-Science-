```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x,y,c='r',alpha=0.5,marker='X')
plt.xlabel('This is my x axis')
plt.ylabel('This is my y axis')
plt.title('x vs y')
plt.show()
```

```python
x = np.linspace(1,10,100)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x,y,c='r')
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y,'--r')
plt.xlabel('x')
plt.ylabel('y=sin(x)')
plt.title('Sine Graph')
plt.grid()
plt.show()
```

```python
x = ['a','b','c','d','e']
y = np.random.rand(len(x))
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(6,4))
plt.bar(x,y,color='b')
plt.xlabel('Categorical Data')
plt.ylabel('Numerical Data')
plt.title('Bar Graph')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.barh(x,y,color='g')
plt.xlabel('Numerical Data')
plt.ylabel('Categorical Data')
plt.title('BarH Graph')
plt.grid()
plt.show()
```

```python
x = [1,2,3,4,4,5,5,5,6,1,2,7,8,6,7,9,9,10,3]
```

```python
plt.hist(x,rwidth=0.8,color='b')
plt.xlabel('Unique Values')
plt.xlabel('Frequency')
plt.title('Histogram example')
plt.savefig('test.jpg')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection='3d')
ax.scatter(x,y,z,c=y)
plt.xlabel('x')
plt.ylabel('y')
#ax.set_zlabel('Z-Axis')
plt.title('3D Graph')
plt.show()
```
