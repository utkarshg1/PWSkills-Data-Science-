# Covariance and Correlation - Utkarsh Gaikwad

```python
import seaborn as sns
```

```python
df = sns.load_dataset('healthexp')
df.head()
```

```python
df.info()
```

```python
# Covariance
df.cov(numeric_only=True)
```

```python
# Correlation
cor = df.corr(numeric_only=True,method='pearson')
cor
```

```python
sns.heatmap(cor,annot=True,fmt='.4f')
```

```python
cor_sp = df.corr(numeric_only=True,method='spearman')
cor_sp
```

```python
sns.heatmap(cor_sp,annot=True,fmt='.4f')
```
